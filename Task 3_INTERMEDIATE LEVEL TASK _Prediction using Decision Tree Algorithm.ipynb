### Importing Libraries ###

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.tree import plot_tree
from sklearn import tree
```

### Reading Data ###

```python
data = datasets.load_iris()
s_data = pd.DataFrame(data.data,columns=data.feature_names)
s_data['target'] = data.target
s_data.head()

```

```python
#spliting data
fc = [x for x in s_data.columns if x!="target"]
x = s_data[fc]
y = s_data["target"]
X_train ,X_test,Y_train,Y_test = train_test_split(x,y,random_state=100,test_size=0.30)
```

```python
#dispaly data
print("X_Train:",X_train.shape)
print("X_Test:",X_test.shape)
print("Y_Train:",Y_train.shape)
print("Y_Test:",Y_test.shape)
```

```python
# Model Creation
classifier = DecisionTreeClassifier()

# Model Training
classifier.fit(X_train, Y_train)

# Predicting
Y_pred = classifier.predict(X_test)
```

```python
# Comparing Actual and Predicted
newdata = pd.DataFrame({"Actual":Y_test,"Predicted":Y_pred})
newdata.head(10)
```

```python
# Testing the accuracy of model
accuracy_score(Y_test,Y_pred)
```

```python
s_data.head()
```

```python
# plotting the classifier tree
fn = ["Sepal length", "Sepal width", "Petal length", "Petal width"]
cn = ["Setosa","Versicolor","Virginica"]
plt.figure(figsize=(10,10))
plot_tree(classifier,feature_names=fn,class_names=cn,filled=True)
```

```python
# Final Result for all x and y
classifierX = DecisionTreeClassifier().fit(x,y)
plt.figure(figsize=(15,15))
tree = plot_tree(classifierX,feature_names=fn,class_names=cn,filled=True)
```
